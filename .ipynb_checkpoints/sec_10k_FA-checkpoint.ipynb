```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("Sec_financial_report.xlsx")
df.head()
```

```python
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
print(df)
```

# Trends and Observations


Revenue Trends:
Apple: Revenue declined in both 2023 (-4.56%) and 2022 (-6.39%).

Microsoft: Revenue declined significantly in 2023 (-13.55%) and slightly in 2022 (-6.44%).

Tesla: Revenue declined sharply in 2021 (-33.93%) and moderately in 2022 (-15.82%).

Net Income Trends:
Apple: Net Income increased in both 2023 (+3.48%) and 2022 (+2.89%).

Microsoft: Net Income declined sharply in 2023 (-17.90%) but increased slightly in 2022 (+0.52%).

Tesla: Net Income declined sharply in both 2021 (-55.16%) and 2022 (-15.94%).

Key Insights
Apple: Despite declining revenue, Apple managed to grow its Net Income, indicating improved cost management or higher profitability per unit of revenue.

Microsoft: Both Revenue and Net Income declined in 2023, suggesting challenges in maintaining profitability during this period.

Tesla: Tesla experienced significant declines in both Revenue and Net Income, particularly in 2021, indicating potential operational or market challenges.

# we will be calculating aditional metrics that indicate the financial leverage of a company
profit margin which is net income divided by total revenue *100
Asset turnover ratio which is Total revenue divided by total assets 
debt-to-equity ratio which is total liabilities divided by(total assets - total liabilities)

```python
df.head(5)
```

```python
columns_to_convert = ['Total Revenue', 'Net Income', 'Total Assets', 'Total Liabilities', 'Cash Flow from operating activities']
```

```python
df['Profit Margin (%)'] = (df['Net Income'] / df['Total Revenue']) * 100

# Calculate Asset Turnover Ratio
df['Asset Turnover Ratio'] = df['Total Revenue'] / df['Total Assets']

# Calculate Debt-to-Equity Ratio
df['Debt-to-Equity Ratio'] = df['Total Liabilities'] / (df['Total Assets'] - df['Total Liabilities'])
```

```python
df.head()
```

```python
print(df)
```

```python
df.columns = [col +"(Millions)" if col in columns_to_convert
             else col
             for col in df.columns]
```

```python
# Visualizations
plt.figure(figsize=(15, 10))

# Plot 1: Revenue and Net Income Trends
plt.subplot(2, 2, 1)
for company in df['Company'].unique():
    company_data = df[df['Company'] == company]
    plt.plot(company_data['Year'], company_data['Total Revenue(Millions)'], label=f'{company} Revenue', marker='o')
    plt.plot(company_data['Year'], company_data['Net Income(Millions)'], label=f'{company} Net Income', linestyle='--', marker='x')
plt.title('Revenue and Net Income Trends')
plt.xlabel('Year')
plt.ylabel('Amount (Millions)')
plt.legend()
plt.grid()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for a cleaner look
sns.set(style="whitegrid")

# Create the figure
plt.figure(figsize=(15, 8))

# Plot 1: Revenue and Net Income Trends
for company in df['Company'].unique():
    company_data = df[df['Company'] == company]
    
    # Plot Revenue
    plt.plot(company_data['Year'], company_data['Total Revenue(Millions)'], 
             label=f'{company} Revenue', marker='o', linestyle='-', linewidth=2, markersize=8)
    
    # Plot Net Income
    plt.plot(company_data['Year'], company_data['Net Income(Millions)'], 
             label=f'{company} Net Income', marker='x', linestyle='--', linewidth=2, markersize=8)

# Add title and labels
plt.title('Revenue and Net Income Trends (2019-2024)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=14, labelpad=10)
plt.ylabel('Amount (Millions)', fontsize=14, labelpad=10)

# Add annotations for key data points
for company in df['Company'].unique():
    company_data = df[df['Company'] == company]
    
    # Annotate Revenue peaks
    max_revenue = company_data['Total Revenue(Millions)'].max()
    max_revenue_year = company_data.loc[company_data['Total Revenue(Millions)'] == max_revenue, 'Year'].values[0]
    plt.annotate(f'{max_revenue:.2f}M', 
                 xy=(max_revenue_year, max_revenue), 
                 xytext=(max_revenue_year, max_revenue + 10), 
                 fontsize=10, color='black', ha='center', 
                 arrowprops=dict(facecolor='black', shrink=0.05))
    
    # Annotate Net Income peaks
    max_net_income = company_data['Net Income(Millions)'].max()
    max_net_income_year = company_data.loc[company_data['Net Income(Millions)'] == max_net_income, 'Year'].values[0]
    plt.annotate(f'{max_net_income:.2f}M', 
                 xy=(max_net_income_year, max_net_income), 
                 xytext=(max_net_income_year, max_net_income + 10), 
                 fontsize=10, color='black', ha='center', 
                 arrowprops=dict(facecolor='black', shrink=0.05))

# Customize legend
plt.legend(title='Company', title_fontsize=12, fontsize=12, bbox_to_anchor=(1.05, 1), loc='upper left')

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for a cleaner look
sns.set(style="whitegrid")

# Create a figure with subplots for each company
fig, axes = plt.subplots(3, 1, figsize=(15, 18))

# Define a color palette for consistency
colors = sns.color_palette("husl", 3)

# Loop through each company and plot
for i, company in enumerate(df['Company'].unique()):
    company_data = df[df['Company'] == company]
    
    # Plot Revenue
    axes[i].plot(company_data['Year'], company_data['Total Revenue(Millions)'], 
                 label='Revenue', marker='o', linestyle='-', linewidth=2, markersize=8, color=colors[0])
    
    # Plot Net Income
    axes[i].plot(company_data['Year'], company_data['Net Income(Millions)'], 
                 label='Net Income', marker='x', linestyle='--', linewidth=2, markersize=8, color=colors[1])
    
    # Add title and labels
    axes[i].set_title(f'{company} - Revenue and Net Income Trends (2019-2024)', fontsize=14, fontweight='bold', pad=10)
    axes[i].set_xlabel('Year', fontsize=12, labelpad=10)
    axes[i].set_ylabel('Amount (Millions)', fontsize=12, labelpad=10)
    
    # Add annotations for key data points
    max_revenue = company_data['Total Revenue(Millions)'].max()
    max_revenue_year = company_data.loc[company_data['Total Revenue(Millions)'] == max_revenue, 'Year'].values[0]
    axes[i].annotate(f'Max Revenue: {max_revenue:.2f}M', 
                     xy=(max_revenue_year, max_revenue), 
                     xytext=(max_revenue_year, max_revenue + 5), 
                     fontsize=10, color=colors[0], ha='center', 
                     arrowprops=dict(facecolor=colors[0], shrink=0.05))
    
    max_net_income = company_data['Net Income(Millions)'].max()
    max_net_income_year = company_data.loc[company_data['Net Income(Millions)'] == max_net_income, 'Year'].values[0]
    axes[i].annotate(f'Max Net Income: {max_net_income:.2f}M', 
                     xy=(max_net_income_year, max_net_income), 
                     xytext=(max_net_income_year, max_net_income + 5), 
                     fontsize=10, color=colors[1], ha='center', 
                     arrowprops=dict(facecolor=colors[1], shrink=0.05))
    
    # Customize legend
    axes[i].legend(title='Metrics', title_fontsize=12, fontsize=12, loc='upper left')
    
    # Add grid for better readability
    axes[i].grid(True, linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python

# Set seaborn style for a cleaner look
sns.set(style="whitegrid")

# Define a color palette for consistency
colors = {
    'Revenue': '#1f77b4',  # Blue
    'Net Income': '#ff7f0e',  # Orange
    'Profit Margin': '#2ca02c',  # Green
    'Asset Turnover': '#9467bd',  # Purple
    'Debt-to-Equity': '#d62728'  # Red
}

# Create a figure with subplots for each company
fig, axes = plt.subplots(3, 4, figsize=(20, 18))

# Loop through each company and plot
for i, company in enumerate(df['Company'].unique()):
    company_data = df[df['Company'] == company]
    
    # Plot 1: Revenue and Net Income Trends
    axes[i, 0].plot(company_data['Year'], company_data['Total Revenue(Millions)'], 
                    label='Revenue', marker='o', linestyle='-', linewidth=2, markersize=8, color=colors['Revenue'])
    axes[i, 0].plot(company_data['Year'], company_data['Net Income(Millions)'], 
                    label='Net Income', marker='x', linestyle='--', linewidth=2, markersize=8, color=colors['Net Income'])
    axes[i, 0].set_title(f'{company} - Revenue and Net Income Trends', fontsize=14, fontweight='bold', pad=10)
    axes[i, 0].set_xlabel('Year', fontsize=12, labelpad=10)
    axes[i, 0].set_ylabel('Amount (Millions)', fontsize=12, labelpad=10)
    axes[i, 0].legend(title='Metrics', title_fontsize=12, fontsize=12, loc='upper left')
    axes[i, 0].grid(True, linestyle='--', alpha=0.7)
    
    # Plot 2: Profit Margin Trends
    axes[i, 1].plot(company_data['Year'], company_data['Profit Margin (%)'], 
                    label='Profit Margin', marker='o', linestyle='-', linewidth=2, markersize=8, color=colors['Profit Margin'])
    axes[i, 1].set_title(f'{company} - Profit Margin Trends', fontsize=14, fontweight='bold', pad=10)
    axes[i, 1].set_xlabel('Year', fontsize=12, labelpad=10)
    axes[i, 1].set_ylabel('Profit Margin (%)', fontsize=12, labelpad=10)
    axes[i, 1].legend(title='Metric', title_fontsize=12, fontsize=12, loc='upper left')
    axes[i, 1].grid(True, linestyle='--', alpha=0.7)
    
    # Plot 3: Asset Turnover Ratio
    axes[i, 2].plot(company_data['Year'], company_data['Asset Turnover Ratio'], 
                    label='Asset Turnover', marker='o', linestyle='-', linewidth=2, markersize=8, color=colors['Asset Turnover'])
    axes[i, 2].set_title(f'{company} - Asset Turnover Ratio', fontsize=14, fontweight='bold', pad=10)
    axes[i, 2].set_xlabel('Year', fontsize=12, labelpad=10)
    axes[i, 2].set_ylabel('Ratio', fontsize=12, labelpad=10)
    axes[i, 2].legend(title='Metric', title_fontsize=12, fontsize=12, loc='upper left')
    axes[i, 2].grid(True, linestyle='--', alpha=0.7)
    
    # Plot 4: Debt-to-Equity Ratio
    axes[i, 3].plot(company_data['Year'], company_data['Debt-to-Equity Ratio'], 
                    label='Debt-to-Equity', marker='o', linestyle='-', linewidth=2, markersize=8, color=colors['Debt-to-Equity'])
    axes[i, 3].set_title(f'{company} - Debt-to-Equity Ratio', fontsize=14, fontweight='bold', pad=10)
    axes[i, 3].set_xlabel('Year', fontsize=12, labelpad=10)
    axes[i, 3].set_ylabel('Ratio', fontsize=12, labelpad=10)
    axes[i, 3].legend(title='Metric', title_fontsize=12, fontsize=12, loc='upper left')
    axes[i, 3].grid(True, linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
